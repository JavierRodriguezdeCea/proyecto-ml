# TABLA ENTRENAMIENTO



```python
tabla1 =  pd.read_csv("https://raw.githubusercontent.com/data-bootcamp-v4/data/main/sales.csv", sep=',')
tabla1
```

```python
tabla1.columns = tabla1.columns.str.lower()
```

```python
# Convertir la columna Date a datetime
tabla1['date'] = pd.to_datetime(tabla1['date'])

# Crear las nuevas columnas year y month
tabla1['year'] = tabla1['date'].dt.year
tabla1['month'] = tabla1['date'].dt.month

# Reordenar las columnas para que year y month estén después de Day_of_week
cols = tabla1.columns.tolist()
cols.insert(cols.index('day_of_week') + 1, cols.pop(cols.index('month')))
cols.insert(cols.index('day_of_week') + 2, cols.pop(cols.index('year')))
tabla1 = tabla1[cols]

# Mostrar el dataframe actualizado
tabla1
```

```python
tabla1 = tabla1[tabla1['open'] != 0]
```

```python
tabla1 = tabla1.drop(columns=['date', 'true_index', "open"])
```

```python
tabla1
```

```python
tabla1['state_holiday'] = tabla1['state_holiday'].apply(lambda x: 1 if x != '0' else 0)
```

```python
tabla1
```

```python
tabla1.dtypes
```

```python
# Cambiar el tipo de dato de las columnas especificadas
tabla1['store_id'] = tabla1['store_id'].astype('category')
tabla1['day_of_week'] = tabla1['day_of_week'].astype('category')
tabla1['month'] = tabla1['month'].astype('category')
tabla1['year'] = tabla1['year'].astype('category')

# Verificar los cambios
tabla1.dtypes
```

```python
tabla1
```

```python
non_numeric_columns = tabla1.select_dtypes(include=['object']).columns
numeric_columns = tabla1.select_dtypes(include=['int64']).columns
```

```python

df_dummies = pd.get_dummies(tabla1, columns=non_numeric_columns)

```

## TRAIN TEST SPLIT

#### Quick EDA

```python
grid = {"n_estimators": [int(x) for x in np.linspace(start = 200, stop = 2000, num = 10)],
        "estimator__max_leaf_nodes": [int(x) for x in np.linspace(start = 500, stop = 3000, num = 10)],
        "estimator__max_depth":[int(x) for x in np.linspace(10, 110, num = 11)]}
```

```python
ada_reg = AdaBoostRegressor(DecisionTreeRegressor())

model = RandomizedSearchCV(estimator = ada_reg, param_distributions = grid, n_iter = 10, cv = 5, n_jobs = -1)
```

```python
features = df_dummies.drop(columns = ["sales"])
target = df_dummies["sales"]
```

```python
X_train, X_test, y_train, y_test = train_test_split(features, target, test_size = 0.20, random_state=0)
```

```python
normalizer = MinMaxScaler()

normalizer.fit(X_train)
```

```python
X_train_norm = normalizer.transform(X_train)

X_test_norm = normalizer.transform(X_test)
```

```python
X_train_norm = pd.DataFrame(X_train_norm, columns = X_train.columns)
X_test_norm = pd.DataFrame(X_test_norm, columns = X_test.columns)
```

```python
model.fit(X_train_norm,y_train)
```

```python
model.best_params_
```

```python
best_model = model.best_estimator_
```

```python
pred = best_model.predict(X_test_norm)

print("MAE", mean_absolute_error(pred, y_test))
print("RMSE", mean_squared_error(pred, y_test,))
print("R2 score", best_model.score(X_test_norm, y_test))
```
